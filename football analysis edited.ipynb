## Remontada indicator 

A Remontada is the act of wining after which we calculate by a different points system where "losing at the first half  (by one goal difference) and then wining is 2 points (adding a goal in the difference = extra point example losing 1-0 first half then winning is 2 points but losing 2-0 first half then winning is 3 points)" and " losing at the first half and then draw is 1 point(the difference property is also applied)" (an extra point if the remontada was in a away game)

```python
remontada = {team: 0 for team in team_names}
for home_team, away_team, ftr, fthg, ftag, htr, hthg, htag in zip(matches.HomeTeam, matches.AwayTeam, matches.FTR, matches.FTHG, matches.FTAG, matches.HTR, matches.HTHG, matches.HTAG):
    if htr == "A" and ftr == "H":
        if htag - hthg == 1:
            remontada[home_team] += 2
        elif htag - hthg == 2:
            remontada[home_team] += 3
        elif htag - hthg >= 3:
            remontada[home_team] += 4
            
    if htr == "H" and ftr == "A":
        if hthg - htag == 1:
            remontada[away_team] += 3
        elif hthg - htag == 2:
            remontada[away_team] += 4
        elif hthg - htag >= 3:
            remontada[away_team] += 5
            
    if htr == "A" and ftr == "D":
        if htag - hthg == 1:
            remontada[home_team] += 1
        elif htag - hthg == 2:
            remontada[home_team] += 2
        elif htag - hthg >= 3:
            remontada[home_team] += 3
            
    if htr == "H" and ftr == "D":
        if hthg - htag == 1:
            remontada[away_team] += 2
        elif hthg - htag == 2:
            remontada[away_team] += 3
        elif hthg - htag >= 3:
            remontada[away_team] += 4
```

```python

team_names_list = list(remontada.keys())

remontada_points_list = list(remontada.values())

y_pos = np.arange(len(team_names_list))

fig, ax = plt.subplots(figsize=(10, 8))

bars = ax.barh(y_pos, remontada_points_list)

ax.set_xlabel('Remontada Points')
ax.set_ylabel('Team')
ax.set_title('Remontada Points for Each Team')
ax.set_yticks(y_pos)
ax.set_yticklabels(team_names_list)

# add values inside each bar
for i, remontada_points in enumerate(remontada_points_list):
    ax.text(remontada_points + 0.5, i, str(remontada_points), ha='left', va='center')

plt.show()
```

## Now lets find which team has the most efficient attack?

This can be determine by calculating the ratio of scored goals to total shots(the deference between this and the goal counts one is efficiency which is an important criteria when evaluating attackers because the ultimate goal is scoring with the least possible amount of tries)

```python
team_shots={team : 0 for team in team_names}
for home_team,away_team,home_shots,away_shots in zip(matches.HomeTeam,matches.AwayTeam,matches.HS,matches.AS):
    team_shots[home_team]=team_shots[home_team]+home_shots
    team_shots[away_team]=team_shots[away_team]+away_shots
```

```python
team_attack_eff={}
for team in team_shots:
    team_attack_eff[team] =round(team_shots[team] / goal_counts[team],3)
```

```python
team_attack_eff_df = pd.DataFrame.from_dict(team_attack_eff, orient='index', columns=['Attack Efficiency'])

# sort the DataFrame by attack efficiency in descending order
team_attack_eff_df = team_attack_eff_df.sort_values(by='Attack Efficiency', ascending=False)

# create a table using Seaborn
sns.set_style('whitegrid')
fig, ax = plt.subplots(figsize=(10, 12))
table = sns.heatmap(team_attack_eff_df, cmap='Reds', annot=True, fmt='.3f', cbar=False, linewidths=0.5, ax=ax)
table.invert_yaxis()
ax.set_ylabel('Team', fontsize=12)
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
ax.set_title('Average shots needed to score a goal (smaller numbers are better)', fontsize=18)
plt.show()
```
